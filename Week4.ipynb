```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
print('=============================') 
print(' Gamma Correction pada citra ') 
print('=============================') 
try: 
 gamma = int(input('Masukkan nilai Gamma: ')) 
except ValueError: 
 print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK/Week4/female.tiff')
gamma_image = np.zeros(original.shape, original.dtype)
#----------------------------------------------#
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      gamma_image[y,x,c] = np.clip(255 * pow(original[y,x,c]/255, (1/gamma)), 0, 255)

final_frame = cv.hconcat((original, gamma_image))
cv2_imshow(final_frame)

```

```python
print('======================') 
print(' Bit Depth pada Citra ')
print('======================') 
try:
  bit_depth = int(input('Masukkan nilai Bit Depth: '))
except ValueError:
  print('Error, not a number')

bit = 2**bit_depth
level = 255/(bit-1)

bit_depth_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      bit_depth_image[y,x,c] = np.clip(round(original[y,x,c] / level) * level, 0, 255)

final_frame = cv.hconcat((original, bit_depth_image))
cv2_imshow(final_frame)
```

```python
import glob
from math import log10, sqrt
```

```python
print('===============================')
print(' Average Denoising 5 Citra ')
print('===============================')

img_amt = 5

def PSNR(ori, compressed):
    mse = np.mean((ori - compressed) ** 2)
    if(mse == 0):
      return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse))
    return psnr, mse

ori = cv.imread('/content/drive/MyDrive/PCVK/Week4/galaxy.jpg')
denoising_avg = np.zeros(ori.shape, ori.dtype)

cv_img = []
for i in glob.glob('/content/drive/MyDrive/PCVK/Week4/noises/*.jpg'):
    n = cv.imread(i)
    cv_img.append(n)

denoising_avg = np.mean(cv_img[:img_amt], axis=0).astype(np.uint8)

psnr_value, mse_value = PSNR(ori, denoising_avg)
print(f"Nilai PSNR: {psnr_value:.2f} dB")
print(f"MSE: {mse_value:.2f}")

final_frame = cv.hconcat((ori, denoising_avg))
cv2_imshow(denoising_avg)
```

```python
import glob
import math
```

```python
print('===========================') 
print(' Average Denoising Citra ')
print('===========================') 

original_image = cv.imread('/content/drive/MyDrive/PCVK/Week4/galaxy.jpg')
cv_img = []
for img in glob.glob('/content/drive/MyDrive/PCVK/Week4/noises/*.jpg'): 
  n = cv.imread(img)
  cv_img.append(n)

def PSNR(img1, img2):
  mse = np.mean((img1 - img2) ** 2)
  if(mse == 0): 
    return 100
  max_pixel = 255.0
  psnr = 20 * math.log10(max_pixel / math.sqrt(mse))
  return psnr

try:
  jumlah_citra = int(input('Masukkan jumlah citra di Average : '))
except ValueError:
  print('Error input')

noised_image = np.zeros(cv_img[0].shape)
for i in range (jumlah_citra):
  noised_image = noised_image+cv_img[i]

  
noised_image = noised_image / jumlah_citra

psnr = PSNR(original_image, noised_image)
print(psnr)
cv2_imshow(noised_image)
```

```python
original = cv.imread('/content/drive/MyDrive/PCVK/Week4/couple.tiff')
masking = np.zeros(original.shape, dtype=np.uint8)
masking = cv.circle(masking, (70, 40), 40, (255,255,255), -1)

result = original & masking

masking1 = cv.circle(masking, (190, 50), 40, (255,255,255), -1)

result = original & masking1

final_frame = cv.hconcat((original, masking, result))
cv2_imshow(final_frame)
```
